## Remove Outliers

```python
import numpy as np
import pandas as pd
from sklearn.utils import shuffle

def tukey_halfspace_depth(data, num_projections=500):
    """
    Approximate Tukey’s Halfspace Depth using random projections.

    Args:
        data (numpy.ndarray): Input numeric data (n_samples, n_features).
        num_projections (int): Number of random halfspaces to project onto.

    Returns:
        numpy.ndarray: Depth scores for each point (higher = more central).
    """
    n, d = data.shape
    depths = np.zeros(n)

    for _ in range(num_projections):
        # Generate a random unit vector direction
        direction = np.random.randn(d)
        direction /= np.linalg.norm(direction)

        # Project data onto this direction
        projections = np.dot(data, direction)

        # Compute depth: Minimum rank distance from either side
        sorted_proj = np.sort(projections)
        left_depth = np.searchsorted(sorted_proj, projections, side='left')
        right_depth = n - np.searchsorted(sorted_proj, projections, side='right')
        depths += np.minimum(left_depth, right_depth)

    return depths / num_projections  # Normalize depths

def clean_outliers_tukey_depth(df, exclude_columns=[], depth_threshold=0.05, num_projections=500):
    """
    Removes outliers using Tukey's Approximate Halfspace Depth.

    Args:
        df (pd.DataFrame): The input DataFrame.
        exclude_columns (list): Columns to exclude from depth calculation.
        depth_threshold (float): Minimum depth score for a point to be kept.
        num_projections (int): Number of random projections for approximation.

    Returns:
        pd.DataFrame: The cleaned DataFrame without outliers.
    """
    df_clean = df.drop(columns=exclude_columns, errors='ignore').copy()
    numeric_cols = df_clean.select_dtypes(include=['number']).columns

    if len(numeric_cols) < 2:
        raise ValueError("Tukey's Depth requires at least two numeric columns.")

    # Convert to numpy array
    data_points = df_clean[numeric_cols].values
    depth_scores = tukey_halfspace_depth(data_points, num_projections=num_projections)

    # Keep only high-depth points
    df_clean["Depth"] = depth_scores
    df_clean = df_clean[df_clean["Depth"] >= depth_threshold]  # Remove outliers
    df_clean.drop(columns=["Depth"], inplace=True)

    # Add back the excluded columns
    df_clean = df_clean.join(df[exclude_columns])

    return df_clean

# Example usage:
dummy_columns = ['Status_Developing', 'Year']
plotdf = clean_outliers_tukey_depth(plotdf, exclude_columns=dummy_columns, depth_threshold=0.05, num_projections=500)
```

```python
def tukey_depth_random_hyperplanes(df, num_planes=2000):
    data = df.values
    n_samples, n_features = data.shape
    depths = np.zeros(n_samples)

    for _ in range(num_planes):
        # Generate a random hyperplane (random normal vector)
        normal_vector = np.random.randn(n_features)

        # Project all points onto this hyperplane
        projections = data @ normal_vector

        # Count how many points are in the smaller half-space
        median_projection = np.median(projections)
        depths += (projections < median_projection)

    return depths
plotdf = eudf.drop(columns=['Country'])
# Compute Approximate Tukey Depth
plotdf['TukeyDepth'] = tukey_depth_random_hyperplanes(plotdf)
threshold = np.percentile(plotdf['TukeyDepth'], 5)
plotdf['Outlier'] = plotdf['TukeyDepth'] <= threshold  # Flag outliers
print(plotdf['Outlier'].value_counts())
plotdf=plotdf[plotdf['Outlier']==False]
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd
import statsmodels.api as sm

# Define the predictors (exclude the target variable, 'Life expectancy')
predictors = plotdf.drop(columns=["Life expectancy "])  # Ensure correct column name

def remove_high_vif(dataframe, threshold=10):
    """Removes predictors with VIF > threshold iteratively while keeping the constant."""

    dataframe = dataframe.copy()  # Avoid modifying original data
    dataframe = sm.add_constant(dataframe)  # Add constant term ONCE

    while True:
        # Calculate VIF for each predictor
        vif_data = pd.DataFrame()
        vif_data["Variable"] = dataframe.columns
        vif_data["VIF"] = [variance_inflation_factor(dataframe.values, i) for i in range(dataframe.shape[1])]

        print("\nCurrent VIF Data:\n", vif_data)

        # Exclude 'const' when checking for high VIF predictors
        high_vif_columns = vif_data[(vif_data["VIF"] > threshold) & (vif_data["Variable"] != "const")]["Variable"].tolist()

        # Stop if no high-VIF predictors remain (except 'const')
        if not high_vif_columns:
            break

        # Drop the predictor with the highest VIF (excluding 'const')
        variable_to_remove = vif_data[vif_data["Variable"].isin(high_vif_columns)].sort_values("VIF", ascending=False)["Variable"].iloc[0]
        dataframe = dataframe.drop(columns=[variable_to_remove], errors="ignore")
        print(f"\nDropped {variable_to_remove} due to high VIF.")

    # Return the cleaned DataFrame **without dropping 'const'**
    return dataframe.drop(columns=["const"], errors="ignore")

# Apply VIF filtering
aftervifdf = remove_high_vif(plotdf.drop(columns=["Life expectancy "]))  # Remove high VIF predictors iteratively

print("\nFinal Dataset After VIF Filtering:\n", aftervifdf.head())
```

```python
import statsmodels.api as sm

def backward_elimination(dataframe, target, threshold=0.05):
    """Performs backward elimination based on p-values."""

    X = dataframe.drop(columns=[target])  # Exclude target variable
    y = dataframe[target]  # Target variable

    # Add constant (Intercept)
    X = sm.add_constant(X)

    while True:
        # Fit model
        model = sm.OLS(y, X).fit()

        # Get p-values
        p_values = model.pvalues.iloc[1:]  # Exclude intercept

        # Find max p-value
        max_p_value = p_values.max()

        if max_p_value < threshold:
            break  # Stop when all predictors are significant

        # Remove the predictor with the highest p-value
        feature_to_remove = p_values.idxmax()
        X = X.drop(columns=[feature_to_remove])

        print(f"Removed '{feature_to_remove}' (p = {max_p_value:.4f})")

    return model

# Run Backward Elimination
final_model = backward_elimination(aftervifdf, target="Life expectancy ")

# Print final model summary
print("\nFinal Model Summary:")
print(final_model.summary())
```

```python
from sklearn.preprocessing import StandardScaler

# Select predictor variables from plotdf (before VIF filtering)
X_lasso = plotdf.drop(columns=["Life expectancy "], errors="ignore")

# Standardize the predictors
scaler = StandardScaler()
X_lasso_scaled = scaler.fit_transform(X_lasso)

# Define target variable
y_lasso = plotdf["Life expectancy "]

print("Standardization Complete! Shape of Data:", X_lasso_scaled.shape)
```

```python
from sklearn.linear_model import LassoCV
from sklearn.model_selection import KFold
from sklearn.metrics import mean_squared_error
import numpy as np

# Set up 5-Fold Cross-Validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# Define Lasso Regression with Cross-Validation to find the best alpha (λ)
lasso = LassoCV(alphas=np.logspace(-3, 2, 100), cv=kf, random_state=42, max_iter=1000000).fit(X_lasso_scaled, y_lasso)

# Print the best alpha (regularization strength)
print(f"Best Alpha (λ) chosen by CV: {lasso.alpha_}")

# Compute RMSE using the best alpha
rmse_scores_lasso = []

for train_index, test_index in kf.split(X_lasso_scaled):
    X_train, X_test = X_lasso_scaled[train_index], X_lasso_scaled[test_index]
    y_train, y_test = y_lasso.iloc[train_index], y_lasso.iloc[test_index]

    # Fit Lasso model with optimal alpha
    model_lasso = LassoCV(alphas=[lasso.alpha_], cv=kf, random_state=42).fit(X_train, y_train)
    y_pred_lasso = model_lasso.predict(X_test)

    # Compute RMSE
    mse_lasso = mean_squared_error(y_test, y_pred_lasso)
    rmse_lasso = np.sqrt(mse_lasso)
    rmse_scores_lasso.append(rmse_lasso)

# Compute Mean & Standard Deviation of RMSE
mean_rmse_lasso = np.mean(rmse_scores_lasso)
std_rmse_lasso = np.std(rmse_scores_lasso)

# Print RMSE for Lasso
print(f"Mean RMSE (Lasso Model): {mean_rmse_lasso:.4f}")
print(f"Standard Deviation of RMSE (Lasso Model): {std_rmse_lasso:.4f}")
```
